```python
import cv2
import time
import os

# Verzeichnis für gespeicherte Bilder erstellen, falls es nicht existiert
output_dir = "./captured_images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Webcam öffnen (0 für die Standard-Webcam)
cap = cv2.VideoCapture(0)

# Setze die Auflösung auf 1080p
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1920)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 1080)

# Überprüfen, ob die Webcam geöffnet werden konnte
if not cap.isOpened():
    print("Fehler: Webcam konnte nicht geöffnet werden.")
    exit()

try:
    counter = 0
    while True:
        # Frame von der Webcam aufnehmen
        ret, frame = cap.read()
        
        # Überprüfen, ob das Frame erfolgreich aufgenommen wurde
        if not ret:
            print("Fehler: Frame konnte nicht aufgenommen werden.")
            break
        
        # Frame anzeigen
        cv2.imshow("Webcam Stream", frame)
        
        # Speichern des Bildes mit einem Zähler im Dateinamen
        filename = f"{output_dir}/{counter}.jpg"
        cv2.imwrite(filename, frame)
        print(f"Bild gespeichert als {filename}")
        
        # Eine Sekunde warten
        time.sleep(1)
        counter += 1
        
        # Überprüfen, ob das Fenster geschlossen werden soll
        if cv2.waitKey(1) & 0xFF == ord('q'):
            print("Beenden angefordert.")
            break

except KeyboardInterrupt:
    print("Aufnahme beendet.")

finally:
    # Webcam freigeben und Fenster schließen
    cap.release()
    cv2.destroyAllWindows()
    print("Webcam geschlossen.")
```
